## Feature Analysis - Target Variable 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../00-data/Housing.csv')
```

```python
# Distribution Analysis
print('Target Variable Anaylsis')
print(f'Max: {df['price'].max()}')
print(f'Min: {df['price'].min()}')
print(f'Mean: {df['price'].mean():.4f}')
print(f'Median: {df['price'].median():.4f}')
print(f'S.t.d: {df['price'].std():.4f}')
print(f'Skeness: {df['price'].skew():4f}')
```

```python
plt.figure(figsize=(10,6))
plt.title('Kernel density Estimation (House Price)')
sns.histplot(df['price'], kde=True)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Outiler Analysis
print(f'Q1: {np.percentile(df['price'], 25):}')
print(f'Q3: {np.percentile(df['price'], 75):}')
print(f'IQR: {(np.percentile(df['price'], 75)) - (np.percentile(df['price'], 25))}')
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(data=df['price'], orient='h')
plt.title('Boxplot (House Price)')
plt.xlabel('Price (Millions)')
```

### Conclusion

- Price distribution is Right skew
- Mean is 4,766,729
- Highest Frequency is around 4 million 
- Outiler Uppper bound is around 900k 
- Outiler Lower bound is around 150k
